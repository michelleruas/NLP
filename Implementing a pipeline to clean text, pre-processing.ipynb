# Implementing a pipeline to clean text

Pre-processing text data

1. Remove punctuation
2. Tokenization
3. Remove stopwords
4. Lemmatize/ Stem

```python
import pandas as pd

#Set custom value of how many characters we can see in dataframe when we print it out (default is 50)
pd.set_option('display.max_colwidth', 100)

data = pd.read_csv("SMSSpamCollection.tsv", sep='\t', header = None)
data.columns = ['label', 'body_text']

data.head()
```

```python
#What we are working towards to clean the data 

data_cleaned = pd.read_csv("SMSSpamCollection_cleaned.tsv", sep='\t')
data_cleaned.head()
```

```python
#Remove punctuation


import string
string.punctuation
```

```python
#Function to remove this punctuation, cycle through and throw away punctuation
#Uses list comprehension

def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

data['body_text_clean'] = data['body_text'].apply(lambda x: remove_punct(x))

data.head()
```

```python
#Tokenization

import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

data['body_text_tokenized'] = data['body_text_clean'].apply(lambda x: tokenize(x.lower()))
data.head()
```

```python
#Remove stopwords

import nltk

stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stop(tokenized_list):
    text_nostop = [word for word in tokenized_list if word not in stopword]
    return text_nostop

data['body_text_nostop'] = data['body_text_tokenized'].apply(lambda x: remove_stop(x))

data.head()
```
